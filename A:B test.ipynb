В вашем приложении существует две модели монетизации: первая — ежемесячная платная подписка, которая позволяет пользователям смотреть ленту без рекламы, вторая — демонстрация рекламы для пользователей, которые ещё не оформили подписку.

Вас, как аналитика, просят помочь рассчитать параметры A/B-теста, который позволит проверить эту гипотезу, и проанализировать его результаты.

#### 1.3. Анализ числа регистраций
Одна из важнейших метрик продукта — число зарегистрированных пользователей. Используя исторические данные, визуализируем, как менялось число регистраций в приложении за время его существования. Пользователь считается зарегистрированным только в день совершения регистрации. Таким образом, необходимо проанализировать количество зарегистрированных активных пользователей за каждый день без накопления (аналог DAU, но для регистраций пользователей).

- Агрегируем исторические данные и рассчитаем число уникальных пользователей и число зарегистрированных пользователей для каждого дня наблюдения. Для простоты считаем, что у пользователя в течение дня бывает одна сессия максимум и статус регистрации в течение одного дня не может измениться.

- Построим линейные графики общего числа пользователей и общего числа зарегистрированных пользователей по дням. 

- Построим отдельный линейный график доли зарегистрированных пользователей от всех пользователей по дням.


```python
sessions_history['session_date'] = pd.to_datetime(sessions_history['session_date'])
sessions_history['install_date'] = pd.to_datetime(sessions_history['install_date'])

daily_stats = sessions_history.groupby('session_date').agg(
    total_users=('user_id', 'nunique'),
    registered_users=('user_id', lambda x: x[sessions_history.loc[x.index, 'registration_flag'] == True].nunique())
).reset_index()

daily_stats = daily_stats.sort_values('session_date')

daily_stats['registered_share'] = daily_stats['registered_users'] / daily_stats['total_users'] * 100

plt.figure(figsize=(12, 6))
plt.plot(daily_stats['session_date'], daily_stats['total_users'], 
         marker='o', linewidth=2, markersize=4, label='Все пользователи', color='blue')
plt.plot(daily_stats['session_date'], daily_stats['registered_users'], 
         marker='s', linewidth=2, markersize=4, label='Зарегистрированные пользователи', color='green')

plt.title('Динамика числа пользователей и зарегистрированных пользователей по дням', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Количество пользователей', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=11, loc='best')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()

plt.figure(figsize=(12, 6))
plt.plot(daily_stats['session_date'], daily_stats['registered_share'], 
         marker='o', linewidth=2, markersize=4, color='red')

plt.title('Доля зарегистрированных пользователей от общего числа пользователей по дням', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Доля зарегистрированных пользователей (%)', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.ylim(0, 100)  # Доля в процентах от 0 до 100
plt.xticks(rotation=45)

plt.axhline(y=50, color='gray', linestyle=':', alpha=0.5, label='50%')


plt.legend(['Доля регистраций', '50%'], fontsize=11, loc='best')

plt.tight_layout()


plt.show()

print(f"Период наблюдения: с {daily_stats['session_date'].min().date()} по {daily_stats['session_date'].max().date()}")
print(f"Всего дней наблюдения: {len(daily_stats)}")
print(f"\nСреднее число пользователей в день: {daily_stats['total_users'].mean():.0f}")
print(f"Среднее число зарегистрированных пользователей в день: {daily_stats['registered_users'].mean():.0f}")
print(f"Средняя доля регистраций: {daily_stats['registered_share'].mean():.1f}%")
print(f"\nМаксимальное число пользователей за день: {daily_stats['total_users'].max()} ({daily_stats.loc[daily_stats['total_users'].idxmax(), 'session_date'].date()})")
print(f"Максимальное число регистраций за день: {daily_stats['registered_users'].max()} ({daily_stats.loc[daily_stats['registered_users'].idxmax(), 'session_date'].date()})")
print(f"Максимальная доля регистраций: {daily_stats['registered_share'].max():.1f}% ({daily_stats.loc[daily_stats['registered_share'].idxmax(), 'session_date'].date()})")
```

#### 1.5. Доля пользователей, просмотревших более четырёх страниц
Продуктовая команда продукта считает, что первые сессии, в рамках которых пользователь просмотрел 4 и более страниц, говорят об удовлетворённости контентом и алгоритмами рекомендаций. Этот показатель является важной прокси-метрикой для продукта.

- В датафрейме `sessions_history` создадим дополнительный столбец `good_session`. В него войдёт значение `1`, если за одну сессию было просмотрено 4 и более страниц, и значение `0`, если было просмотрено меньше.

- Построим график со средним значением доли успешных первых сессий от всех первых сессий пользователей. Данные по дням за весь период наблюдения.

```python
sessions_history = pd.read_csv('/datasets/sessions_project_history.csv')
sessions_history['session_date'] = pd.to_datetime(sessions_history['session_date'])
sessions_history['good_session'] = (sessions_history['page_counter'] >= 4).astype(int)

first_sessions = sessions_history[sessions_history['session_number'] == 1].copy()

daily_stats = first_sessions.groupby('session_date').agg(
    total_first_sessions=('session_id', 'count'),
    good_first_sessions=('good_session', 'sum')
).reset_index()

daily_stats['good_session_share'] = daily_stats['good_first_sessions'] / daily_stats['total_first_sessions']
daily_stats = daily_stats.sort_values('session_date')

print(f"Всего первых сессий: {daily_stats['total_first_sessions'].sum():,}".replace(',', ' '))
print(f"Из них успешных (4+ страниц): {daily_stats['good_first_sessions'].sum():,}".replace(',', ' '))
print(f"Общая доля успешных первых сессий: {daily_stats['good_session_share'].mean()*100:.2f}%")
print(f"Медианная доля успешных сессий по дням: {daily_stats['good_session_share'].median()*100:.2f}%")
print(f"\nПериод наблюдения: с {daily_stats['session_date'].min().date()} по {daily_stats['session_date'].max().date()}")
print(f"Всего дней с первыми сессиями: {len(daily_stats)}")

plt.figure(figsize=(14, 7))

plt.plot(daily_stats['session_date'], 
         daily_stats['good_session_share'] * 100,
         marker='o', 
         markersize=5,
         linewidth=2, 
         color='#2E86AB',
         label='Доля успешных первых сессий')

window = 7
if len(daily_stats) >= window:
    rolling_mean = daily_stats['good_session_share'].rolling(window=window, center=True).mean() * 100
    plt.plot(daily_stats['session_date'], 
             rolling_mean, 
             linewidth=3, 
             color='#C73E1D', 
             alpha=0.8,
             linestyle='--',
             label=f'Скользящее среднее ({window} дней)')

overall_mean = daily_stats['good_session_share'].mean() * 100
plt.axhline(y=overall_mean, 
            color='gray', 
            linestyle=':', 
            alpha=0.7, 
            linewidth=2,
            label=f'Общее среднее: {overall_mean:.1f}%')

plt.title('Доля успешных первых сессий пользователей по дням\n(сессии с просмотром 4+ страниц)', 
          fontsize=16, 
          fontweight='bold', 
          pad=20)

plt.xlabel('Дата', fontsize=13, fontweight='bold')
plt.ylabel('Доля успешных первых сессий (%)', fontsize=13, fontweight='bold')

plt.grid(True, alpha=0.3, linestyle='--', linewidth=0.7)
plt.legend(fontsize=12, loc='best')

plt.xticks(rotation=45, ha='right')

min_val = daily_stats['good_session_share'].min() * 100
max_val = daily_stats['good_session_share'].max() * 100
margin = (max_val - min_val) * 0.1
plt.ylim(min_val - margin, max_val + margin)

plt.tight_layout()
plt.show()
```

### 2. Подготовка к тесту
При планировании теста необходимо проделать несколько важных шагов:

- Определиться с целевой метрикой.

- Рассчитать необходимый размер выборки.

- Исходя из текущих значений трафика рассчитать необходимую длительность проведения теста.

#### 2.1. Расчёт размера выборки


```python
from statsmodels.stats.power import NormalIndPower
from statsmodels.stats.proportion import proportion_effectsize

alpha = 0.05
beta = 0.2
power = 1 - beta
p = 0.3
relative_mde = 0.03
absolute_mde = p * relative_mde
effect_size = proportion_effectsize(p, p + absolute_mde)

power_analysis = NormalIndPower()

sample_size = power_analysis.solve_power(
    effect_size=effect_size,
    power=power,
    alpha=alpha,
    ratio=1
)

print(f"Необходимый размер выборки для каждой группы: {int(sample_size)}")
```

#### 2.2. Расчёт длительности A/B-теста


```python
from math import ceil

sessions_history['session_date'] = pd.to_datetime(sessions_history['session_date'])

avg_daily_users = sessions_history.groupby('session_date')['user_id'].nunique().mean()

sample_size = 41040

test_duration = ceil(sample_size * 2 / avg_daily_users)

print(f"Среднее количество уникальных пользователей в день: {avg_daily_users:.0f}")
print(f"Рассчитанная длительность A/B-теста при текущем уровне трафика в {avg_daily_users:.0f} пользователей в день составит {test_duration} дней")
```

### 3. Мониторинг А/В-теста

#### 3.1. Проверка распределения пользователей

A/B-тест успешно запущен, и уже доступны данные за первые три дня. На этом этапе нужно убедиться, что всё идёт хорошо: пользователи разделены правильным образом, а интересующие вас метрики корректно считаются.


```python
sessions_test_part = pd.read_csv('/datasets/sessions_project_test_part.csv')
sessions_test_part['session_date'] = pd.to_datetime(sessions_test_part['session_date'])
users_by_group = sessions_test_part.groupby('test_group')['user_id'].nunique()

for group in ['A', 'B']:
    if group in users_by_group.index:
        print(f"Группа {group}: {users_by_group[group]:,} уникальных пользователей".replace(',', ' '))

users_A = users_by_group.get('A', 0)
users_B = users_by_group.get('B', 0)

if users_A > 0:
    percent_diff = 100 * abs(users_A - users_B) / users_A
else:
    percent_diff = 0

print(f"\nПроцентная разница между группами: {percent_diff:.2f}%")

# Оценка качества сплитования
expected_ratio = 0.5
actual_ratio = users_B / (users_A + users_B) if (users_A + users_B) > 0 else 0
print(f"Ожидаемое распределение: 50% / 50%")
print(f"Фактическое распределение: {users_A/(users_A+users_B)*100:.1f}% / {users_B/(users_A+users_B)*100:.1f}%")
print(f"Отклонение от ожидаемого: {abs(actual_ratio - expected_ratio)*100:.2f}%")

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

ax1 = axes[0]
groups = ['A', 'B']
users = [users_A, users_B]
colors = ['#2E86AB', '#A23B72']
bars = ax1.bar(groups, users, color=colors, edgecolor='black', alpha=0.8)
ax1.set_title('Количество уникальных пользователей по группам', fontsize=12, fontweight='bold')
ax1.set_xlabel('Тестовая группа', fontsize=11)
ax1.set_ylabel('Количество пользователей', fontsize=11)
ax1.grid(axis='y', alpha=0.3, linestyle='--')

for bar, user_count in zip(bars, users):
    ax1.text(bar.get_x() + bar.get_width()/2., bar.get_height() + max(users)*0.01,
             f'{int(user_count):,}'.replace(',', ' '),
             ha='center', va='bottom', fontsize=10)

ax2 = axes[1]
ax2.pie(users, labels=groups, autopct='%1.1f%%', colors=colors, startangle=90, explode=(0.02, 0.02))
ax2.set_title('Распределение пользователей (доли)', fontsize=12, fontweight='bold')

ax3 = axes[2]
diff_percent = (abs(users_A - users_B) / users_A) * 100
colors_diff = ['green' if diff_percent < 5 else 'orange' if diff_percent < 10 else 'red']
ax3.bar(['Процентная разница'], [diff_percent], color=colors_diff[0], edgecolor='black', alpha=0.8)
ax3.axhline(y=5, color='green', linestyle='--', alpha=0.7, label='Приемлемый уровень (5%)')
ax3.axhline(y=10, color='orange', linestyle='--', alpha=0.7, label='Пограничный уровень (10%)')
ax3.set_title(f'Процентная разница между группами\n({diff_percent:.2f}%)', fontsize=12, fontweight='bold')
ax3.set_ylabel('Разница (%)', fontsize=11)
ax3.set_ylim(0, max(diff_percent + 5, 15))
ax3.grid(axis='y', alpha=0.3, linestyle='--')
ax3.legend(fontsize=9)

plt.suptitle('Проверка распределения пользователей в A/B-тесте\n(данные за первые три дня)', 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

#### 3.2. Проверка пересечений пользователей

#### 4.2 Формулировка нулевой и альтернативной гипотез. Определение целевой, прокси- и барьерных метрик


Нулевая гипотеза (H₀):

Новый рекомендательный алгоритм не приводит к изменению доли успешных сессий (сессий с просмотром 4+ страниц) по сравнению со старым алгоритмом.
Математически: Доля успешных сессий в группе B (новый алгоритм) равна доле успешных сессий в группе A (старый алгоритм).

Альтернативная гипотеза (H₁):

Новый рекомендательный алгоритм приводит к увеличению доли успешных сессий (сессий с просмотром 4+ страниц) по сравнению со старым алгоритмом.
Математически: Доля успешных сессий в группе B больше, чем в группе A.

Прокси-метрики: Среднее количество страниц за сессию (page_counter), Доля пользователей, вернувшихся на следующий день (Retention Day 1)
Барьерные: Доля пользователей, завершивших сессию после 1 страницы, Количество сессий на пользователя в день, Средняя ошибка рекомендаций

#### 4.3. Сравнение доли успешных первых сессий

Перейдем к анализу ключевой метрики — доле успешных первых сессий.

Используйте созданный на первом шаге задания столбец `good_session` и рассчитайте долю успешных первых сессий для выборок A и B, а также разницу в этом показателе. Полученный вывод отобразите на экране.

```python
first_sessions = sessions_test[sessions_test['session_number'] == 1].copy()

group_A_first = first_sessions[first_sessions['test_group'] == 'A']
group_B_first = first_sessions[first_sessions['test_group'] == 'B']

share_A = group_A_first['good_session'].mean() * 100
share_B = group_B_first['good_session'].mean() * 100

diff = share_B - share_A

print("Доля успешных первых сессий (good_session = 1)")
print(f"\nГруппа A (контрольная, старый алгоритм):")
print(f"  - Всего первых сессий: {len(group_A_first):,}".replace(',', ' '))
print(f"  - Успешных сессий (4+ страниц): {group_A_first['good_session'].sum():,}".replace(',', ' '))
print(f"  - Доля успешных сессий: {share_A:.2f}%")

print(f"\nГруппа B (тестовая, новый алгоритм):")
print(f"  - Всего первых сессий: {len(group_B_first):,}".replace(',', ' '))
print(f"  - Успешных сессий (4+ страниц): {group_B_first['good_session'].sum():,}".replace(',', ' '))
print(f"  - Доля успешных сессий: {share_B:.2f}%")

print(f"  Абсолютная разница: {diff:+.2f} процентных пункта")
print(f"  Относительное изменение: {diff/share_A*100:+.2f}%")

if diff > 0:
    print(f"\nДоля успешных первых сессий в группе B выше, чем в группе A на {diff:.2f} п.п.")
elif diff < 0:
    print(f"\nДоля успешных первых сессий в группе B ниже, чем в группе A на {abs(diff):.2f} п.п.")
else:
    print(f"\nДоля успешных первых сессий в группах A и B одинакова.")
```

#### 4.4. Насколько статистически значимо изменение ключевой метрики


```python
from statsmodels.stats.proportion import proportions_ztest

sessions_test = pd.read_csv('/datasets/sessions_project_test.csv')

sessions_test['good_session'] = (sessions_test['page_counter'] >= 4).astype(int)

first_sessions = sessions_test[sessions_test['session_number'] == 1].copy()

group_A = first_sessions[first_sessions['test_group'] == 'A']
group_B = first_sessions[first_sessions['test_group'] == 'B']

success_A = group_A['good_session'].sum()
success_B = group_B['good_session'].sum()
n_A = len(group_A)
n_B = len(group_B)

share_A = success_A / n_A * 100
share_B = success_B / n_B * 100
diff = share_B - share_A

count = [success_B, success_A]
nobs = [n_B, n_A]

z_stat, p_value = proportions_ztest(count, nobs, alternative='larger')

print(f"\nГруппа A (контроль): {share_A:.2f}% ({success_A} из {n_A})")
print(f"Группа B (тест): {share_B:.2f}% ({success_B} из {n_B})")
print(f"Разница (B - A): {diff:+.2f} процентных пункта")

print(f"Результаты статистического теста (z-test для пропорций, односторонний):")
print(f"  z-статистика: {z_stat:.4f}")
print(f"  p-value: {p_value:.6f}")
print(f"  Уровень значимости (alpha): 0.05")

if p_value < 0.05:
    print(f"\n p-value ({p_value:.6f}) < 0.05")
    print(f"→ Различие между группами является СТАТИСТИЧЕСКИ ЗНАЧИМЫМ.")
    print(f"→ Новый алгоритм (группа B) показывает более высокую долю успешных первых сессий.")
    print(f"→ Рекомендуется отклонить нулевую гипотезу в пользу альтернативной.")
else:
    print(f"\n p-value ({p_value:.6f}) >= 0.05")
    print(f"→ Различие между группами НЕ ЯВЛЯЕТСЯ статистически значимым.")
    print(f"→ Новый алгоритм не показал статистически значимого улучшения.")
    print(f"→ Нет оснований отклонять нулевую гипотезу.")
```
